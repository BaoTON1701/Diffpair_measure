```python
# icvce_RT = np.genfromtxt('R&TBiCMOS_IHP/IcVce_77K_1.txt')
import pandas as pd
IcVce_RT = pd.read_csv('R&TBiCMOS_IHP/IcVce_RT.csv')
IcVce_CT1 = pd.read_csv('R&TBiCMOS_IHP/icVce_77K_Ib500nto2250nA.csv')
IcVce_CT2 = pd.read_csv('R&TBiCMOS_IHP/icVce_77K_Ib1to8uA.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_segmented_curves(df, x_col, y_col, num_curves, ax=None,
                          title=None, xlabel=None, ylabel=None, **plot_kwargs):
    """
    Plots data from concatenated sweeps onto a specified or new matplotlib axes.

    Args:
        df (pd.DataFrame): DataFrame with the plotting data.
        x_col (str): Column name for the x-axis.
        y_col (str): Column name for the y-axis.
        num_curves (int): The number of distinct curves concatenated in the data.
        ax (matplotlib.axes.Axes, optional): The axes object to plot on.
            If None, a new figure and axes are created. Defaults to None.
        title (str, optional): Title for the plot. Overwrites existing title if ax is provided.
        xlabel (str, optional): Label for the x-axis.
        ylabel (str, optional): Label for the y-axis.
        **plot_kwargs: Additional keyword arguments passed to plt.plot().

    Returns:
        tuple: (fig, ax) The matplotlib Figure and Axes objects.
    """
    # --- 1. Validation ---
    total_points = len(df)
    if total_points % num_curves != 0:
        print("Error: Data cannot be split evenly.")
        return None, None
    if x_col not in df.columns or y_col not in df.columns:
        print("Error: Column not found.")
        return None, None

    points_per_curve = total_points // num_curves
    
    # --- 2. Handle Axes Creation (The Key Change) ---
    if ax is None:
        # If no axes are provided, create a new plot
        fig, ax = plt.subplots(figsize=(12, 8))
    else:
        # If axes are provided, use them and get their parent figure
        fig = ax.get_figure()

    # --- 3. Plotting Loop ---
    for i in range(num_curves):
        start_index = i * points_per_curve
        end_index = start_index + points_per_curve
        curve_data = df.iloc[start_index:end_index]
        ax.plot(curve_data[x_col], curve_data[y_col], **plot_kwargs)

    # --- 4. Customization ---
    # Only set labels/title if they are explicitly provided
    if title:
        ax.set_title(title)
    if xlabel:
        ax.set_xlabel(xlabel)
    if ylabel:
        ax.set_ylabel(ylabel)
    ax.grid(True)

    return fig, ax
```
