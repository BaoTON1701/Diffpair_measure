```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat300K_1 = np.genfromtxt('R&TBiCMOS_IHP/gummel_filter1_300k.txt')
dat77K_2 = np.genfromtxt('R&TBiCMOS_IHP/gummel_filter2_77k.txt')
dat300K_2  = np.genfromtxt('R&TBiCMOS_IHP/gummel_300k.txt')
```

```python
# --- Plotting Code ---
fig, ax1 = plt.subplots(figsize=(10, 7)) # Slightly larger figure for better readability
ax2 = ax1.twinx()

ax1.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,1]*1e3, color = 'red', lw=3, label = r'$I_B$')
ax1.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,2]*1e3, color = 'blue', lw=3, label = r'$I_C$')
ax2.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,3], color ='green',linestyle = 'dashed', lw=3, label = r'$\beta$')
# --- Axis Labels and Titles ---
ax1.set_xlabel("$V_{BE} = V_{CE}$ [V]", fontsize=16) # Added units
ax1.set_ylabel("Current [mA]", fontsize=16, color='black') # Explicitly set color for clarity

ax2.set_ylabel("Current Gain ($\\beta$)", fontsize=16, color='green') # Added LaTeX for beta, set color
ax2.tick_params(axis='y', labelcolor='green') # Make ax2 tick labels green to match the line

# --- Grid ---
ax1.grid(True, which="both", ls="-", color='0.7', linewidth=0.5) # Both major and minor grid lines, light grey
# You could also add a grid specifically to ax2 if you want it to be different
# ax2.grid(True, which="both", ls=":", color='0.8', linewidth=0.3)

# --- Ticks and Appearance ---
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10) # Optional: if you want minor tick labels
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10) # Optional

# Add minor ticks, which are very useful for log plots
ax1.minorticks_on()
ax2.minorticks_on()


# --- Legends ---
# Combine legends from both axes into one
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='lower right', fontsize=12, frameon=True, fancybox=True, shadow=True) # Unified legend, placed for clarity

# --- Title ---
fig.suptitle("BJT Gummel Plot: Current vs. Voltage and Current Gain @300K", fontsize=18, y=1.02) # More descriptive title, slightly higher

# --- Layout and Saving ---
fig.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
```
