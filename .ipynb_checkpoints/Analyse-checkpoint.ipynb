```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_segmented_curves(df, x_col, y_col, num_curves, ax=None,
                          curves_to_plot=None, title=None, xlabel=None,
                          ylabel=None, **plot_kwargs):
    """
    Plots selected or all curves from concatenated sweeps.

    Args:
        df (pd.DataFrame): DataFrame with the plotting data.
        x_col (str): Column name for the x-axis.
        y_col (str): Column name for the y-axis.
        num_curves (int): The TOTAL number of distinct curves available in the data.
        ax (matplotlib.axes.Axes, optional): The axes object to plot on.
        curves_to_plot (list of int, optional): A list of specific curve indices
            to plot (e.g., [0, 2, 5]). If None, all curves are plotted.
        title (str, optional): Title for the plot.
        xlabel (str, optional): Label for the x-axis.
        ylabel (str, optional): Label for the y-axis.
        **plot_kwargs: Additional keyword arguments for plt.plot().
    """
    # --- 1. Validation ---
    total_points = len(df)
    if total_points % num_curves != 0:
        print("Error: Data cannot be split evenly.")
        return None, None
    points_per_curve = total_points // num_curves

    # --- 2. Determine which curves to iterate over ---
    if curves_to_plot is None:
        # If no specific curves are requested, plot all of them
        plot_indices = range(num_curves)
    else:
        # If a list is provided, validate it and use it
        if any(i >= num_curves or i < 0 for i in curves_to_plot):
            print(f"Error: `curves_to_plot` has an invalid index. Use indices from 0 to {num_curves - 1}.")
            return None, None
        plot_indices = curves_to_plot

    # --- 3. Setup for plotting ---
    if ax is None:
        fig, ax = plt.subplots(figsize=(12, 8))
    else:
        fig = ax.get_figure()

    label = plot_kwargs.pop('label', None)
    is_first_plot = True  # Use a flag to apply label only to the first curve plotted

    # --- 4. Plotting Loop ---
    for i in plot_indices:  # Loop over only the selected indices
        start_index = i * points_per_curve
        end_index = start_index + points_per_curve
        curve_data = df.iloc[start_index:end_index]

        # Apply label only to the first curve in our selection
        if is_first_plot:
            ax.plot(curve_data[x_col], curve_data[y_col], label=label, **plot_kwargs)
            is_first_plot = False # The label is now used, don't use it again
        else:
            ax.plot(curve_data[x_col], curve_data[y_col], **plot_kwargs)

    # --- 5. Customization ---
    if title: ax.set_title(title)
    if xlabel: ax.set_xlabel(xlabel)
    if ylabel: ax.set_ylabel(ylabel)
    ax.grid(True)

    return fig, ax
```
