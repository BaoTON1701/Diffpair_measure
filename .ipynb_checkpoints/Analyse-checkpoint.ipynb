```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat300K_1 = np.genfromtxt('R&TBiCMOS_IHP/gummel_filter1_300k.txt')
dat77K_2 = np.genfromtxt('R&TBiCMOS_IHP/gummel_filter2_77k.txt')
dat300K_2  = np.genfromtxt('R&TBiCMOS_IHP/gummel_300k.txt')
```

```python
# --- Plotting Code ---
fig, ax1 = plt.subplots(figsize=(10, 7)) # Slightly larger figure for better readability
ax2 = ax1.twinx()

ax1.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,1]*1e3, color = 'red', lw=3, label = r'$I_B$')
ax1.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,2]*1e3, color = 'blue', lw=3, label = r'$I_C$')
ax2.semilogy(dat300K_1[1:-1,0], dat300K_1[1:-1,3], color ='green',linestyle = 'dashed', lw=3, label = r'$\beta$')
# --- Axis Labels and Titles ---
ax1.set_xlabel("$V_{BE} = V_{CE}$ [V]", fontsize=16) # Added units
ax1.set_ylabel("Current [mA]", fontsize=16, color='black') # Explicitly set color for clarity

ax2.set_ylabel("Current Gain ($\\beta$)", fontsize=16, color='green') # Added LaTeX for beta, set color
ax2.tick_params(axis='y', labelcolor='green') # Make ax2 tick labels green to match the line

# --- Grid ---
ax1.grid(True, which="both", ls="-", color='0.7', linewidth=0.5) # Both major and minor grid lines, light grey
# You could also add a grid specifically to ax2 if you want it to be different
# ax2.grid(True, which="both", ls=":", color='0.8', linewidth=0.3)

# --- Ticks and Appearance ---
ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='minor', labelsize=10) # Optional: if you want minor tick labels
ax2.tick_params(axis='both', which='major', labelsize=12)
ax2.tick_params(axis='both', which='minor', labelsize=10) # Optional

# Add minor ticks, which are very useful for log plots
ax1.minorticks_on()
ax2.minorticks_on()


# --- Legends ---
# Combine legends from both axes into one
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='lower right', fontsize=12, frameon=True, fancybox=True, shadow=True) # Unified legend, placed for clarity

# --- Title ---
fig.suptitle("BJT Gummel Plot: Current vs. Voltage and Current Gain @300K", fontsize=18, y=1.02) # More descriptive title, slightly higher

# --- Layout and Saving ---
fig.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
```

```python


# --- Plotting Code ---
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8), gridspec_kw={'height_ratios': [2, 2]})

# Create the twin axis for the gain plot on the right
ax2 = ax1.twinx()

# Plot Ic and Ib (300K) on the left axis (ax1)
l1, = ax1.semilogy(dat300K_1[:,0], dat300K_1[:,2], color='red', lw=2.5, label=r'$I_C$ (300K)')
l2, = ax1.semilogy(dat300K_1[:,0], dat300K_1[:,1], color='red', lw=2.5, linestyle='--', label=r'$I_B$ (300K)')

# Plot Ic and Ib (77K) on the left axis (ax1)
l3, = ax1.semilogy(dat77K_2[:,0], dat77K_2[:,2], color='deepskyblue', lw=2.5, label=r'$I_C$ (77K)')
l4, = ax1.semilogy(dat77K_2[:,0], dat77K_2[:,1], color='deepskyblue', lw=2.5, linestyle='--', label=r'$I_B$ (77K)')

# Plot Beta (300K and 77K) on the right axis (ax2)
l5, = ax2.semilogy(dat300K_1[:,0], dat300K_1[:,3], color='black', lw=2.5, label=r'$\beta$ (300K)')
l6, = ax2.semilogy(dat77K_2[:,0], dat77K_2[:,3], color='black', lw=2.5, linestyle='--', label=r'$\beta$ (77K)')


# --- Styling and Labels ---

# Set axis labels and colors
ax1.set_xlabel("$V_{BE} = V_{CE}$ [V]", fontsize=16)
ax1.set_ylabel("$I_C, I_B$ [A]", fontsize=16, color='black')
ax2.set_ylabel(r"$\beta$ [A/A]", fontsize=16, color='black')

# Set axis limits to match the reference plot
ax1.set_xlim(0.3, 1.0)
ax1.set_ylim(1e-13, 1e-1) # 100fA to 100mA
ax2.set_ylim(1, 1000)

# Set tick parameters
ax1.tick_params(axis='y', which='major', labelsize=12)
ax2.tick_params(axis='y', which='major', labelsize=12)
ax1.tick_params(axis='x', which='major', labelsize=12)

# Add grid
ax1.grid(True, which="both", ls="-", color='0.7', linewidth=0.5)
ax1.minorticks_on()

# Add shaded regions using axvspan
ax1.axvspan(0.38, 0.45, color='grey', alpha=0.5, zorder=0)
ax1.axvspan(0.8, 0.9, color='grey', alpha=0.5, zorder=0)

# --- Combined Legend ---
# Collect handles and labels from both axes and create a single legend
handles = [l1, l2, l3, l4, l5, l6]
labels = [h.get_label() for h in handles]
ax1.legend(handles, labels, loc='center right', fontsize=12)

# --- Layout and Saving ---
fig.tight_layout() # Adjust layout to prevent labels from overlapping
plt.savefig('plot/gummel_plot_combined.png', dpi=300)
plt.show()
```
